## Review on Manipulating matrices

It is straightforward to create a Matrix using Numpy. Let us consider the following as a examples:

$$
A = \begin{pmatrix}
5 & 6 & 2\\
4 & 7 & 19\\
0 & 3 & 12
\end{pmatrix}
$$

$$
B = \begin{pmatrix}
14 & -2 & 12\\
4 & 4 & 5\\
5 & 5 & 1
\end{pmatrix}
$$


First, similarly to Sympy, we need to import Numpy:

````python
import numpy as np
````

Now we can define $A$:

````python
A = np.matrix([[5, 6, 2],
               [4, 7, 19],
               [0, 3, 12]])
````

````python
A
````

````python
B = np.matrix([[14, -2, 12],
               [4, 4, 5],
               [5, 5, 1]])
````

````python
B
````

We can obtain the following straightforwardly:

- $5A$ (or any other scalar multiple of $A$);
- $A ^ 3$ (or any other exponent of $A$);
- $A + B$;
- $A - B$;
- $AB$

````python
5 * A
````

````python
A ** 3
````

````python
A + B
````

````python
A - B
````

````python
A * B
````

````python
A@B
````

## Solving Matrix equations

Solve system of homogenoeus linear equation

x + y - z = 0  
-2x - 3y + z = 0  
x - 4y + 2z = 0

We can use Numpy to (efficiently) solve large systems of equations of the form:

````python
A = np.array([[ 1,  1, -1],
              [-2, -3, -1],
              [ 1, -4, 2]])

b = np.array([0, 0, 0])

solution = np.linalg.solve(A, b)
print("Solution:", solution)
````

Generally, the system of linear equation can be written into the form
$$Ax=b$$

Let us illustrate that with:

$$
A = \begin{pmatrix}
5 & 6 & 2\\
4 & 7 & 19\\
0 & 3 & 12
\end{pmatrix}
$$

$$
b = \begin{pmatrix}
-1\\
2\\
1
\end{pmatrix}
$$

````python
A = np.matrix([[5, 6, 2],
               [4, 7, 19],
               [0, 3, 12]])
b = np.matrix([[-1], [2], [1]])
````

We use the `linalg.solve` command:

````python
x = np.linalg.solve(A, b)
x
````

We can verify our result:

````python
A * x
````

## Matrix inversion and determinants

Computing the inverse of a matrix is straightforward:

````python
Ainv = np.linalg.inv(A)
Ainv
````

We can verify that $A^{-1}A=\mathbb{I}$:

````python
A * Ainv
````

The above might not look like the identity matrix but if you look closer you see that the diagonals are all `1` and the off diagonals are a **very** small number (which from a computer's point of view is `0`).

````python
import numpy as np
import sympy as sy

A = np.array(A, dtype=float)
Ainv = np.array(Ainv, dtype=float)

AA = sy.Matrix(A.tolist()) * sy.Matrix(Ainv.tolist())
````

To calculate the determinant:

````python
np.linalg.det(A)
````

## Rank and Nullity

````python
A = np.array([[1, -1, 3],
              [5, -4, -4],
              [7, -6, 2]])

rank = np.linalg.matrix_rank(A)
nullity = A.shape[1] - rank

print("Rank of A:", rank)
print("Nullity of A:", nullity)
````

````python
A = np.array([[1, 3, 4, 3],
              [3, 9, 12, 9],
              [1, 3, 4, 1]])

rank = np.linalg.matrix_rank(A)
nullity = A.shape[1] - rank

print("Rank of A:", rank)
print("Nullity of A:", nullity)
````
